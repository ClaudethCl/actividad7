```python
#Importamos las librerías para trabajar
#Siendo una de ellas Seaborn.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

```

```python
#Creamos un data frame con el archivo de datos
df0 = pd.DataFrame( pd.read_csv("meteo-nogal-09.csv", engine="python") )
df0.head()
```

```python
#Del archivo podrán notar que trae muchas columnas sin nombre y con datos vacíos
#Pidamosle a python que nos diga el tipo de variables 
df0.dtypes
```

```python
#En análisis de datos, gran parte del tiempo se invertirá en extraer los datos útiles para poder trabajar con ellos.
#Para este trabajo, por favor elimina todo aquello que no brinda información, sin destruir el archivo de datos original.
#Por lo tanto eliminaremos aquello que diga Unnamed.
df0=df0.drop(df0.columns[df0.columns.str.contains('unnamed:',case = False)],axis = 1)
df0.head()
```

```python
#Observamoslos tipos de variables otra vez (*)
df0.dtypes
```

```python
#Creamos una columna que con las dates y las horas te encapsulelas fechas
df0["Fecha"] = df0["DATE"] + " "+ df0["TIME"]
#Quitamos las columnas date y time
df0.drop( ["DATE","TIME"], axis=1, inplace=True )
df0.head()
```

```python
#El renglon de las unidades no es necesario, para lo que queremos hacer así que lo quitaremos
#Haremos que el ídice vuelva a empezar desde cero a pesar de haberle quitado un renglón
df0 = df0.drop(df0.index[0]).reset_index(drop=True)
df0.head()
```

```python
#Como vimos anteiormente (en (*)) teníamos casi puras variables de tipo objeto , traduciremos éstasvariables a float's64
df0[df0.columns[0:14]] = df0[df0.columns[0:14]].apply(pd.to_numeric, errors='coerce')
df0.dtypes
```

```python
#Creamos un data frame df1 que contenga a las correlaciones (las cualesse obtienen usando a la función .corr)
df1 = df0.corr(method='pearson', min_periods=1)
df1.head()
```

```python
#Usaremos a la bibliotécad e SeaBorn para graficar a las correlaciones
grafica = sns.heatmap(df1,robust=True,square=True,annot=False)
grafica.set_title("Gráfica de correlaciones en SeaBorn") 
fig = grafica.get_figure()

#sns.heatmap(df1, mask=8, cmap=9, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
